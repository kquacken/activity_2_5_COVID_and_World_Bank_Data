```python
dvlp_reshape = (dvlp_countries
            >>select(~X.Series_Code)
            >>gather('yr','numbers',columns_from('_2018_YR2018'),add_id = True)
            >>filter_by(X._ID <1330) #gets rid of NAN column
            >>select(~X._ID)
            >>spread(X.Series_Name,X.numbers)
            >>mutate(yr = X.yr.str.replace('_2018_YR2018','2018'))
            >>mutate(yr = X.yr.str.replace('_2019_YR2019','2019'))
            >>mutate(yr = X.yr.astype('datetime64'))
            >>mutate(yr = X.yr.dt.year)
            )
dvlp_reshape
```
